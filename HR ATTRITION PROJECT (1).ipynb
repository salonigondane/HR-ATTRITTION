# standardization 

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x1 = pd.DataFrame(ss.fit_transform(x[con]),columns = con)
x2 = pd.get_dummies(x[cat])
xnew = x1.join(x2)
```

# part 1 : predicting "monthly rate"

```python
RS = 47   # random state
MD = 2    # max depth
dtr_ML = 49   # min samples leaf
dtr_MS = 49   # min samples split

rfr_ML = 48   # min samples leaf
rfr_MS = 48   # min samples split


# 1.LinearRegression

from sklearn.linear_model import LinearRegression, Lasso, Ridge
lr = LinearRegression()
model = lr.fit(xtrain,ytrain)
pred_test = model.predict(xtest)
pred_train = model.predict(xtrain)

from sklearn.metrics import mean_absolute_error
print('LinearRegression')
print('Test  :   ',round(mean_absolute_error(ytest,pred_test),3))
print('Train :   ',round(mean_absolute_error(ytrain,pred_train),3))
print('----------------------------------------------------')

# 2. DecisionTreeRegressor

from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state=RS)
dtr_model = dtr.fit(xtrain,ytrain)
pred_test =dtr_model.predict(xtest)
pred_train =dtr_model.predict(xtrain)

print('DecisionTreeRegressor')
print('Test  :   ',round(mean_absolute_error(ytest,pred_test),3))
print('Train :   ',round(mean_absolute_error(ytrain,pred_train),3))
print('----------------------------------------------------')

# dtr -- max_depth
dtr = DecisionTreeRegressor(random_state=RS, max_depth=MD)
dtr_model = dtr.fit(xtrain,ytrain)
pred_test=dtr_model.predict(xtest)
pred_train=dtr_model.predict(xtrain)

print('DecisionTreeRegressor : max_depth ')
print('Test  :   ',round(mean_absolute_error(ytest,pred_test),3))
print('Train :   ',round(mean_absolute_error(ytrain,pred_train),3))
print('----------------------------------------------------')

# dtr -- min_samples_leaf
dtr = DecisionTreeRegressor(random_state=RS, min_samples_leaf=dtr_ML)
dtr_model = dtr.fit(xtrain,ytrain)
pred_test=dtr_model.predict(xtest)
pred_train=dtr_model.predict(xtrain)

print('DecisionTreeRegressor: min_samples_leaf')
print('Test  :   ',round(mean_absolute_error(ytest,pred_test),3))
print('Train :   ',round(mean_absolute_error(ytrain,pred_train),3))
print('----------------------------------------------------')


# dtr -- min_samples_split
dtr = DecisionTreeRegressor(random_state=RS, min_samples_split=dtr_MS)
dtr_model = dtr.fit(xtrain,ytrain)
pred_test=dtr_model.predict(xtest)
pred_train=dtr_model.predict(xtrain)

print('DecisionTreeRegressor: min_samples_split')
print('Test  :   ',round(mean_absolute_error(ytest,pred_test),3))
print('Train :   ',round(mean_absolute_error(ytrain,pred_train),3))
print('----------------------------------------------------')


# 3. Lasso

ls = Lasso()
tg=[]

m=0.005
for i in range(0,200,1):
    m=m+0.0001
    m=round(m,3)
    tg.append(m)
    
tp={'alpha':tg}

from sklearn.model_selection import GridSearchCV
cv = GridSearchCV(ls,tp, scoring='neg_mean_absolute_error', cv=4)
cvmodel = cv.fit(xtrain,ytrain)
p = cvmodel.best_params_['alpha']

ls = Lasso(alpha=p)
lsmodel = ls.fit(xtrain,ytrain)
pred_test = lsmodel.predict(xtest)
pred_train = lsmodel.predict(xtrain)

print('Lasso')
print('Test  :   ',round(mean_absolute_error(ytest,pred_test),3))
print('Train :   ',round(mean_absolute_error(ytrain,pred_train),3))
print('----------------------------------------------------')


# 4. Ridge

rr = Ridge()
cv = GridSearchCV(rr,tp,scoring='neg_mean_absolute_error', cv=4)
cvmodel = cv.fit(xtrain,ytrain)
n = cvmodel.best_params_['alpha']

rr = Ridge(alpha=n)
rrmodel= rr.fit(xtrain,ytrain)
pred_test = rrmodel.predict(xtest)
pred_train = rrmodel.predict(xtrain)

print('Ridge')
print('Test  :   ',round(mean_absolute_error(ytest,pred_test),3))
print('Train :   ',round(mean_absolute_error(ytrain,pred_train),3))
print('----------------------------------------------------')


# 5. AdaBoostRegressor

from sklearn.ensemble import AdaBoostRegressor
abr = AdaBoostRegressor(dtr,random_state=RS)

tp = {'n_estimators' : range(2,50,1)}
cv = GridSearchCV(abr,tp,scoring="neg_mean_absolute_error",cv=4)
cvmodel = cv.fit(xtrain,ytrain)
ne = cvmodel.best_params_['n_estimators']

abr = AdaBoostRegressor(dtr, random_state=RS, n_estimators = ne)
abrmodel = abr.fit(xtrain,ytrain)
pred_test = abrmodel.predict(xtest)
pred_train = abrmodel.predict(xtrain)

print('AdaBoostRegressor')
print('Test  :   ',round(mean_absolute_error(ytest,pred_test),3))
print('Train :   ',round(mean_absolute_error(ytrain,pred_train),3))
print('----------------------------------------------------')

# 6. RandomForestRegressor

from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state=RS)
rfrmodel = rfr.fit(xtrain,ytrain)
pred_test = rfrmodel.predict(xtest)
pred_train = rfrmodel.predict(xtrain)

print('RandomForestRegressor')
print('Test  :   ',round(mean_absolute_error(ytest,pred_test),3))
print('Train :   ',round(mean_absolute_error(ytrain,pred_train),3))
print('----------------------------------------------------')

# rfr ----max_depth
rfr = RandomForestRegressor(random_state=31, max_depth=MD)
rfrmodel = rfr.fit(xtrain,ytrain)
pred_test = rfrmodel.predict(xtest)
pred_train = rfrmodel.predict(xtrain)

print('RandomForestRegressor: max_depth')
print('Test  :   ',round(mean_absolute_error(ytest,pred_test),3))
print('Train :   ',round(mean_absolute_error(ytrain,pred_train),3))
print('----------------------------------------------------')


# rfr ----min_samples_leaf
rfr = RandomForestRegressor(random_state=RS, min_samples_leaf=rfr_ML)
rfrmodel = rfr.fit(xtrain,ytrain)
pred_test = rfrmodel.predict(xtest)
pred_train = rfrmodel.predict(xtrain)

print('RandomForestRegressor: min_samples_leaf')
print('Test  :   ',round(mean_absolute_error(ytest,pred_test),3))
print('Train :   ',round(mean_absolute_error(ytrain,pred_train),3))
print('----------------------------------------------------')


# rfr ----min_samples_split
rfr = RandomForestRegressor(random_state=RS, min_samples_split=rfr_MS)
rfrmodel = rfr.fit(xtrain,ytrain)
pred_test = rfrmodel.predict(xtest)
pred_train = rfrmodel.predict(xtrain)

print('RandomForestRegressor: min_samples_split')
print('Test  :   ',round(mean_absolute_error(ytest,pred_test),3))
print('Train :   ',round(mean_absolute_error(ytrain,pred_train),3))
print('----------------------------------------------------')


# 7. KNeighborsRegressor

from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor()

tp = {'n_neighbors': range(2,50,1)}
cv = GridSearchCV(knr,tp, scoring='neg_mean_absolute_error', cv=4)
cvmodel = cv.fit(xtrain,ytrain)
nn = cvmodel.best_params_['n_neighbors']

knr = KNeighborsRegressor(n_neighbors= nn)
knrmodel = knr.fit(xtrain,ytrain)
pred_test = knrmodel.predict(xtest)
pred_train = knrmodel.predict(xtrain)

print('KNeighborsRegressor')
print('Test  :   ',round(mean_absolute_error(ytest,pred_test),3))
print('Train :   ',round(mean_absolute_error(ytrain,pred_train),3))
print('----------------------------------------------------')
```
